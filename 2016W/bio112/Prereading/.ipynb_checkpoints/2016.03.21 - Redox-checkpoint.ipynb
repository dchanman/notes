# Chapter 7 Introduction
**Cellular respiration** is a series of chemical reactions that convert energy stored in fuel molecules into a chemical form that can be readily used by cells.

# 7.1 Cellular Respiration

## Cellular respiration occurs in four stages

Four stages:
1. **Glycolysis**: `glucose -> pyruvate`
2. `pyruvate -> acetyl-CoA + CO2`
3. **Citric acid cycle**: `acetyl-CoA` further broken down, more $CO_2$ released
    * these first three steps creates ATP and **electron carriers**
4. **Oxydative phosphorylation**: electron carriers donate high energy electrons to **electron transport chains (respiratory chains)**
    * Oxygen is the last acceptor, so oxygen is consumed and water is produced.
    
In eukaryotes:
* glycosis in cytoplasm
* citric acid cycle + oxidative phosphorylation in mitochondria
* electron transport chain in proteins inside mitochondrial membrane

In eukaryotes:
* reactions in cytoplasm
* electron transport chain in plasma membrane

## Cellular respiration involves a series of redox reactions

**Oxidation-reduction reactions** are used to store and release chemical energy.

**Oxidation** loses electrons  
**Reduction** gains electrons  
Both always occur in a single redox reaction

In many biological systems, electrons are not completely transferred. Rather, *redox involves increases/decreases in electron density* where electrons are *partially lost/gained*

For example, a `C-C` bond becoming a `C=O` bond is an decreased electron density (oxidation). A `O=O` bond becoming a `H-O` bond is an increase in electron density (reduction)

**Oxidizing agents** are **electon aceptors**  
**Reducing agents** are **electron donors**  
Note how these definitions are kinda flipped from redox above.

$C_6H_{12}O_6$ is reduced to $CO_2$  
$CO_2$ is oxidized to $H_2O$

## Chemical energy is stored in reduced molecules such as carbohydrates and lipids
Carbs and lipids have a lot of energy because `C-C` and `C-H` bonds have high potential energy.

Energy is released gradually in cellular respiration. Glucose is oxidized slowly through many reactions.

Electron carriers contain chemical energy in the form of high-energy electrons, which can be harnessed to produce ATP.

## Electron carriers transport high-energy electrons
Electron carriers have two forms: (oxidized/reduced)
* **nicotinamide adenine dinucleotide** ($NAD^+ / NADH$)
* **flavin adenine dinucleotide** ($FADH/FADH_2$)

Reduced form has high potential energy.

## ATP is generated by substrate-level phosphorylation and oxidative phosphorylation
**Substrate-level phosphorylation** is a a way of generating ATP
* phosphorylated organic molecule directly transfers phosphate to ADP

**Oxidative phosphorylation** is the common way of generating ATP
* ATP generated indirectly
* through reduction of electron carriers, transfer of high-energy electrons
* $ADP + P_i \rightarrow ATP$
