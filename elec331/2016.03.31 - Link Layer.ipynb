# Random Access Protocols

When a node has packets to send
* transmit at full channel data rate *R*
* no *a priori* coordination among nodes

**Collisions** occur when 2+ nodes transmit

**Random Access MAC protocols** specifies
* how to detect collisions
* how to recover from collisions

Example random access MAC:
* slotted ALOHA
* ALOHA
* CSMA

## Slotted ALOHA

**Assumptions**
* all frames are same size
* time divided into equal timeslots
* nodes start transmitted only at slot beginning
* nodes are synchronized
* all nodes can detect a collision
* proven to be **37%** efficient ($\frac{1}{e}$)

**Operation**
* if no collision: transmit in the next slot as well
* if collision: "flip a coin" and retransmit next slot on success

**Pros**
* simple
* highly decentralized
* single node can continuously transmit at full rate

**Cons**
* collisions
* idle slots
* clock synchronization required
* nodes need to detect collision in less time than transmitting packet

## Pure ALOHA

* No slots
* More collisions
* proven to be **18%** efficient ($\frac{1}{2e}$)
* This is okay as long as you only have a few nodes in your network. This is simple!

## CSMA/CD

**CSMA** Carrier sense multiple access:
* listen before transmit, if busy, wait. if free, gogogo

**CD** Collision detection
* collisions detected within short time
* colliding transmissions aborted, reducing channel wastage
* implementation
    * easy in wired LANs: simply measure signal strengths and compare
    * difficult in wireless LANs: reeived signal strength overwhelmed by local transmission strength
* similar to a **human polite conversationalist**

Efficiency:
$\frac{1}{1 + 5\frac{t_{prop}}{t_{trans}}}$

# MAC Addresses and ARP

32-bit IP address
* network-layer address for interface used in layer 3 (network layer) forwarding

MAC (or LAN/physical/ethernet) address
* used locally to get frame from one interface to another physically connected interface (same network)
* 48-bit (6 bytes) MAC address burned in NIC ROM

Assigned by IEEE, manufacturers buy MAC adress bulks

MAC never changes, unlike IP

## ARP: Address Resolution Protocol
 > How to determine interface's MAC address knowing its IP?
 
 **ARP table**
 * IP/MAC address mapping for some LAN nodes
 * TTL: time after address mapping will be forgotten
 
 ARP is plug and play!
