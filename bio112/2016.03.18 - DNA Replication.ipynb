# Role of Telomerase

Synthesis of the lagging strand results in a shorter DNA strand. This is because we have to have the RNA primer at the very end of the template strand. Removing the RNA primer results in a small section of the daughter strand that was not replicated.

Telomerase **extends the template strand** so that primase can bind beyond the entire strand. Then DNA polymerase can replicate the entire template, including the ends.

---

There is current research that believes that aging is the result of this shortening of chromosomes.

However, tumour cells typically take over telomerase and use for continued, unhindered growth.

# Errors in DNA Replication

### Clicker
> T/F: An error in DNA replication is the same as a mutation

False.

> T/F: DNA damage is the same as a mutation

False: DNA damage is **not** the same as a mutation.

## Recall: Mutations
Mutations occur:
1. Randomly in nature caused by **unrepaired mistakes in DNA that makes it through two rounds of replication**
2. DNA damage from environmental exposure to mutagens, which influence the likelihood of replication errors

Recall that mutations are not all detrimental. Natural mutation allows for diversity.

### Nobel Prize in Chemistry 2015
Awarded to Lindahl, Modrich, Sancar for *mechanistic studies of DNA repair*.

### Mistakes Happen
Rate occurs 1:100,000 bases.

**Errors can be fixed during replication**
* point mutations
* insertion/deletion mutations
* fixed with ***DNA polymerase proofreading function***

**Missed mutations not corrected by proofreading can be fixed after DNA replication**
* fixed with ***DNA mismatch repair systems***

## DNA Polymerase Proofreading Function
* DNA Polymerase recognizes inappropriate base pairing
* Triggers an enzymatic activity: **exonuclease activity**

> The proofreading response of DNA polymerase is likely triggered by:

*...the addition of the incorrect nucleotide altering the geometry of the DNA.*

Recall that the Watson-Crick base pairings have specific geometries.

## DNA Mismatch Repair Systems
We are focussing on a single repair system. 

1. Repair mechanism is triggered. Mismatched bases in DNA initiate the repair process
2. Backbone cleavage exposes the bases
3. Exonuclease removes nucleotides in the region (including ones that may not be mismatched)
4. Gap is filled, strand is sealed (some sort of DNA polymerase)

## Other types of DNA damage
DNA can be damaged by
* natural cellular processes
    * metabolism byproducts (free radicals, oxidizing agents)
    * addition of chemical groups (methylation, alkylation)
* exposure to mutagens
    * ionizing radiation
    * ultraviolet radiation
    * chemical mutagens
    
On its own, **ligase** is a repair mechanism that preserves the structure of DNA.

There are around 10,000 to 1,000,000 molecular lesions per cell per day. Due to different mechanisms of repair, we can fix this damage and lower the likelihood of mutations.

## Repairing DNA Damage
* **ligase** repairs single strand breaks
* **base excision repair** fixes a single nucleotide
* **nucleotide excision repair** fixes major structural kinks in the DNA, more than one nucleotide is involved

**Example**: UV radiation: creates a covalent bond between thymines. Creates a kink in DNA.

## Summary
* Errors in DNA replication can lead to mutations, but are not all mutations
    * mismatch repair (single, short base pairs)
    * base excision repair (base swapped out)
    * nucleotide excision repair (longer stretches)

Recall: We need to replicate DNA in order to pass genetic information to daughter cells that undergo division
