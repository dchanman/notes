## Glycolysis Phases

### Phase 1 (Energy Investment Phase):
```
               ____________
 Glucose      |            |          Glucose 6-phosphate
    ========> | Glycolysis | ========>
              |            |
 2 ATP        |            |          ADP
    --------> |            | -------->
              |            |              
              |____________|
```
* Notice how we have to use some ATP to start
* Aside: this traps glucose in the cell 
* This is to phosphorylate the glucose molecule

### Phase 2 (Cleavage Phase):
```
Glucose        ____________
6-phosphate   |            |          Glyceraldahyde 3-phosphate
    ========> | Glycolysis | ========>
              |            |
              |            | 
              |            | 
              |            |           
              |____________|
```
* NAD+ to NADH is a redox reaction. NAD+ gets reduced to NADH. These molecules' sole purpose
* Note that CO2 is not yet produced! Our carbons from glucose all exist in pyruvate.

### Phase 3 (Payoff Phase):
```
Glyceraldehyde ____________
3-phosphate   |            |          Pyruvate
    ========> | Glycolysis | ========>
              |            |
 NAD+         |            |          NADH
    --------> |            | -------->
 ADP          |            |          ATP
    --------> |            | -------->
              |            |           
              |____________|
```

### High Energy Intermediates
* NAD+ can accept 2 electons (and reduce to NADH)
* NADH can donate 2 electrons (and oxidize to NAD+)
